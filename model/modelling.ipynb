```python
#postcodes 
nhs_postcodes = duckdb.read_csv('../data/NHSPD_FEB_2024_UK_EXTRACT/Data/*.csv', filename = True)
nhs_postcodes_2 = duckdb.sql("""SELECT column0 AS post_code,
           SPLIT(filename,'/')[-1] as file_name,
           CASE SPLIT(filename,'/')[-1]
WHEN 'feb24y56.csv' THEN 'London Region'
WHEN 'feb24y58.csv' THEN 'South West Region'
WHEN 'feb24y59.csv' THEN 'South East Region'
WHEN 'feb24y60.csv' THEN 'Midlands Region'
WHEN 'feb24y61.csv' THEN 'East of England Region'
WHEN 'feb24y62.csv' THEN 'North West Region'
WHEN 'feb24y63.csv' THEN 'North East and Yorkshire Region'
WHEN 'feb24wsn.csv' THEN 'Wales, Scotland and Northern Ireland'
           ELSE 'Other' END AS nhs_region
           FROM nhs_postcodes """)
duckdb.sql("""SELECT nhs_region, COUNT(*) FROM nhs_postcodes_2 GROUP BY nhs_region""")
duckdb.sql("""SELECT * FROM nhs_postcodes_2 WHERE post_code ILIKE 'TQ2'""")
```

```python
pc_ll_csv = duckdb.read_csv('../data/cleansed/ons_postcodes_latlong.csv')
pc_ll_df = duckdb.sql("""SELECT * FROM pc_ll_csv""")
pc_ll_df.df()
```

```python
export_df = duckdb.sql("""SELECT nhs_postcodes_2.nhs_region, pc_df.PostCode,  ncd_df.*, pc_ll_df.lat, pc_ll_df.long FROM ncd_df 
    LEFT JOIN pc_df ON pc_df.hospitalcode = ncd_df.ODSCode
    LEFT JOIN nhs_postcodes_2 ON REPLACE(TRIM(UPPER(nhs_postcodes_2.post_code)),' ','') = REPLACE(TRIM(UPPER(pc_df.PostCode)),' ','')
    LEFT JOIN pc_ll_df ON TRIM(nhs_postcodes_2.post_code) IN (pcd,pcd2)
        WHERE ncd_df.country = 'ENGLAND'
        AND ncd_df.ODSCode NOT IN ('NTP','NPG','NYR')
                       """)
export_df.df()
```
